```python
# data cleaning
# Quality Metrics Before Cleaning
print("==== Quality Metrics Before Cleaning ====")
print("Total Rows:",len(df))
print("Duplicated Rows:",df.duplicated().sum())
print("Missing Values per column:")
df.isnull().sum()
```

```python
# executing data cleaning pipeline
# remove duplicate rows
df_clean = df.drop_duplicates().copy()
df_clean.duplicated().sum()
```

```python
# clean Gender and Returned Columns case & strip whitespace
df_clean["Gender"] = df_clean["Gender"].astype(str).str.strip().str.upper()
gender_map = {"MALE":"Male","M":"Male","FEMALE":"Female","F":"Female"}
df_clean["Gender"] = df_clean["Gender"].map(gender_map)
df_clean["Gender"].value_counts(dropna=False)
```

```python
df_clean["Returned"] = df["Returned"].astype(str).str.strip().str.upper()
returned_map = {"YES":"Yes","NO":'No'}
df_clean["Returned"] = df_clean["Returned"].map(returned_map)
df_clean["Returned"].value_counts(dropna=False)
```

```python
# check product category and payment method columns
print("Product_Category Value Counts")
df_clean["Product_Category"].value_counts(dropna=False)
```

```python
print("Payment Method Value Counts")
df_clean["Payment_Method"].value_counts(dropna=False)
```

```python
# clean product category and payment method columns
df_clean["Product_Category"] = df_clean["Product_Category"].astype(str).str.strip().str.title()
df_clean["Product_Category"].value_counts(dropna=False) 
```

```python
df_clean["Payment_Method"] = df_clean["Payment_Method"].astype(str).str.strip()
payment_map = {
    "Cash On Delivery":"Cash on Delivery",
    "COD":"Cash on Delivery",
    "Net Banking":"Net Banking",
    "UPI":"UPI",
    "upi":"UPI",
    "Debit Card":"Debit Card",
    "DEBIT CARD":"Debit Card",
    "Credit Card":"Credit Card",
    "credit card":"Credit Card"
}
df_clean["Payment_Method"] = df_clean["Payment_Method"].map(lambda x: payment_map.get(x,x))
df_clean["Payment_Method"].value_counts(dropna=False)
```

```python
# check price and discount columns
print("Price Value Counts")
df_clean["Price"].head()
```

```python
df_clean["Discount"].head() 
```

```python
# clean price & discount columns
df_clean["Price"] = df_clean["Price"].astype(str).str.replace("$","",regex=False).str.strip()
df_clean["Price"] = pd.to_numeric(df_clean["Price"],errors="coerce")
df_clean["Price"].head()
```

```python
# clean discount column
def clean_discount(val):
    if pd.isna(val):
        return 0.0
    val_str = str(val).strip()
    if val_str.endswith("%"):
        try:
            return float(val_str.replace("%",'')) / 100.0
        except:
            return 0.0
    try:
        return float(val_str)
    except:
        return 0.0

df_clean["Discount"] = df_clean["Discount"].apply(clean_discount)
df_clean["Discount"].head()
```

```python
# apply absolute value to handle negative prices
df_clean["Price"] = df_clean["Price"].abs()

# fill missing price values using median
median_prices = df_clean.groupby("Product_Name")["Price"].transform("median")
df_clean["Price"] = df_clean["Price"].fillna(median_prices)
```

```python
# handle missing,negative,quantity values
# calculate implied quantity from sales,price and discount
implied_quantity = (df_clean["Sales"].abs() / (df_clean["Price"] * (1-df_clean["Discount"]))).round()
median_qty = df_clean["Quantity"].median()

if pd.isna(median_qty) or median_qty <= 0:
    median_qty = 3.0

implied_quantity = implied_quantity.fillna(median_qty)
implied_quantity = implied_quantity.replace([np.inf,-np.inf],median_qty)

# fill invalid or missing quantities
invalid_quantity = df_clean["Quantity"].isna() | (df_clean["Quantity"] <=0) | (df_clean["Quantity"] > 50)
df_clean.loc[invalid_quantity,"Quantity"] = implied_quantity[invalid_quantity]
df_clean["Quantity"] = df_clean["Quantity"].astype(int)
```

```python
# recalculate sales column to fix negative values
df_clean["Sales"] = (df_clean["Quantity"] * df_clean["Price"] * (1-df_clean["Discount"])).round(2)
```

```python
# treat profit column anomalies
full_profit = df_clean["Profit"] > df_clean["Sales"]
df_clean.loc[full_profit,"Profit"] = (df_clean["Sales"] * 0.15).round(2)
```

```python
# clean age column
valid_ages = df_clean[(df_clean["Age"] >= 18) & (df_clean["Age"] <= 100)]["Age"]
median_age = valid_ages.median() if len(valid_ages) > 0 else 40.0
df_clean["Age"] = df_clean["Age"].fillna(median_age)
df_clean.loc[(df_clean["Age"] < 18) | (df_clean["Age"] > 100)]["Age"]
df_clean["Age"] = df_clean["Age"].astype(int)
```

```python
# clean rating column(1-5)
median_rating = df_clean[(df_clean["Rating"] >= 1.0) & (df_clean["Rating"] <= 5.0)]["Rating"].median()
if pd.isna(median_rating):
    median_rating = 3.0

df_clean["Rating"] = df_clean["Rating"].fillna(median_rating)
df_clean.loc[(df_clean["Rating"] < 1.0) | (df_clean["Rating"] > 5.0),"Rating"] = median_rating
```

```python
# clean date columns
order_date_dt = pd.to_datetime(df_clean["Order_Date"],errors="coerce")
delivery_date_dt = pd.to_datetime(df_clean["Delivery_Date"],errors="coerce")

# calculate median delivery duration
duration = (delivery_date_dt - order_date_dt).dt.days
median_duration = duration[duration > 0].median()
if pd.isna(median_duration):
    median_duration = 5.0
median_duration = pd.Timedelta(days=int(median_duration))

# missing order date and delivery date values
missing_order_date = order_date_dt.isna() & delivery_date_dt.notna()
order_date_dt[missing_order_date] = delivery_date_dt[missing_order_date] - median_duration

missing_delivery_date = delivery_date_dt.isna() & order_date_dt.notna()
delivery_date_dt[missing_delivery_date] = order_date_dt[missing_delivery_date] + median_duration

# fallbacks if both dates are missing or invalid
both_missing = order_date_dt.isna() & delivery_date_dt.isna()
order_date_dt[both_missing] = pd.to_datetime("2023-01-01")
delivery_date_dt[both_missing] = pd.to_datetime("2023-01-01") + median_duration

# correct physically impossible dates
invalid_duration_date = delivery_date_dt < order_date_dt
delivery_date_dt[invalid_duration_date] = order_date_dt[invalid_duration_date] + median_duration

df_clean["Order_Date"] = order_date_dt
df_clean["Delivery_Date"] = delivery_date_dt
```

```python
# handle customer name
df_clean["Customer_Name"] = df_clean["Customer_Name"].fillna("Unknown")
```

```python
# strip spaces from other string columns
string_cols = ["Customer_Name","City","State","Product_Category","Product_Name","Payment_Method"]

for i in string_cols:
    df_clean[i] = df_clean[i].astype(str).str.strip()
```

```python
# checking quality metrics after cleaning
print("==== Quality Metrics After Cleaning ====")
print("Total Rows:",len(df_clean))
print("Duplicated Rows:",df_clean.duplicated().sum())
print("\nMissing values per column:")
print(df_clean.isnull().sum())
print("\nUnique Product Categories:")
print(df_clean["Product_Category"].unique())
print("\nUnique Payment Methods:")
print(df_clean["Payment_Method"].unique())
print("\nCleaned Numerical Summary:")
df_clean[['Age','Quantity','Sales','Profit','Rating']].describe()
```

```python
# EDA
# 1. Total Sales,Profit and Profit Margin

total_sales = df_clean["Sales"].sum()
total_profit = df_clean["Profit"].sum()

profit_margin = (total_profit / total_sales) * 100
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
# 2. Top 10 Cities by Sales
top_cities = df_clean.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)
top_cities.map(lambda x: f"${x:,.2f}")
```

```python
# 3. Category performance(Sales, Profit, Profit Margin)
category_performance = df_clean.groupby("Product_Category")[["Sales","Profit"]].sum().sort_values(by="Sales",ascending=False)
category_performance["Profit_Margin"] = (category_performance["Profit"] / category_performance["Sales"]) * 100
category_performance
```

```python
# 4. Average Customer Ratings by Category
avg_ratings = df_clean.groupby("Product_Category")["Rating"].mean().sort_values(ascending=False)
avg_ratings.round(2)
```

```python
# 5. Payment Method Distribution
payment_distribution = df_clean["Payment_Method"].value_counts()
payment_distribution
```

```python
# 6. Product Returns Analysis
total_transactions = len(df_clean)
returned_transactions = (df_clean["Returned"] == "Yes").sum()

overall_returned_rate = (returned_transactions / total_transactions) * 100
print(f"Overall Returned Rate: {overall_returned_rate:.2f}%")

# returned rate by category
returned_rate_by_category = df_clean.groupby("Product_Category")["Returned"].apply(lambda x: (x == "Yes").sum() / len(x) * 100).sort_values(ascending=False)
returned_rate_by_category
```

```python
# 7. Monthly Sales Trends
df_clean["Order_Month"] = df_clean["Order_Date"].dt.to_period("M")

monthly_sales_trend = df_clean.groupby('Order_Month')['Sales'].sum()
monthly_sales_trend.map(lambda x: f"${x:,.2f}")
```
